```python
import numpy as np
import pandas as pd

from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv("cleanedDataset.csv")
```

## Encode Categorical Variables

```python
#Label Encoding (turning categorical to numerical)
cat_cols=[]
for i in data.columns:
    if data[i].nunique() <= 5 or data[i].dtype == object:
        cat_cols.append(i)
norm_df=data.copy()

# Converting cols to category
for i in data[cat_cols]:
  data[i] = pd.Categorical(data[i])
  data[i] = data[i].cat.codes
```

```python
data = data.drop(columns=["Unnamed: 0"])
data
```

## SMOTE Sampling technique

```python
cols = list(data.columns)
cols.remove("Attrition")
sampled,target = SMOTE().fit_resample(data[cols],data["Attrition"])
```

```python
newdata = pd.concat([sampled,target], axis=1)
newdata
```

```python
newdata.to_csv('cleanedResampledSMOTEDataset.csv')
```
